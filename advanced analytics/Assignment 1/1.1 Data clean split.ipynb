### Converting the  data type of columns

### Fixing Date columns

```python
df_transactions.loc[:, df_transactions.columns.str.contains("date", case=False)]
```

```python
df_transactions.loc[:, df_transactions.columns.str.contains("date", case=False)].dtypes
```

```python
cols_to_convert = df_transactions.columns[
    df_transactions.columns.str.contains("date", case=False)
]
```

```python
df_transactions[cols_to_convert] = df_transactions[cols_to_convert].apply(
    pd.to_datetime, errors="coerce", format="%Y-%m-%d"
)
```

```python
df_transactions.head()
```

```python
df_transactions.dtypes
```

```python
df_transactions[cols_to_convert].isna().sum()
```

all the date columns are turned to appropriate datetime objects. the formats were consistent, so there are no null values

### Fixing `prod_size`

```python
df_transactions["prod_size"].sample(n=10)
```

```python
pd.to_numeric(df_transactions["prod_size"], errors="coerce").isna().sum()
```

there seems to be a value in the `prod_size` which is not numeric

```python
not_numeric_mask = pd.to_numeric(df_transactions["prod_size"], errors="coerce").isna()
```

```python
df_transactions.loc[not_numeric_mask, "prod_size"]
```

XS is not a valid shoe size, and since it's only one record we can safely drop it

```python
df_transactions = df_transactions[~not_numeric_mask]
```

```python
df_transactions.shape
```

```python
df_transactions["prod_size"] = pd.to_numeric(df_transactions["prod_size"])
```

```python
df_transactions["prod_size"].dtype
```

# Creating train and valid datasets

```python
df_revenues = pd.read_csv("data/customer_clv_train.csv")
```

```python
df_revenues.head()
```

```python
df_revenues_train, df_revenues_valid = train_test_split(df_revenues, test_size=0.2, random_state=7)
```

```python
df_revenues_train
```

```python
df_revenues_train.to_csv("data/train_targets.csv", index=False)
```

```python
df_revenues_valid
```

```python
df_revenues_valid.to_csv("data/valid_targets.csv", index=False)
```

```python
train_cust_ids = df_revenues_train["cust_id"]
```

```python
valid_cust_ids = df_revenues_valid["cust_id"]
```

```python
df_train = df_transactions[df_transactions["cust_id"].isin(train_cust_ids)].copy()
```

```python
df_valid = df_transactions[df_transactions["cust_id"].isin(valid_cust_ids)].copy()
```
